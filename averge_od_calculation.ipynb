```python
# def getAvergedOD(od_df2019,od_df2020,od_df2021,od_df2022,iz_dict):
    
#     avg_od=(od_df2019.merge(
#         od_df2020[['origin','destination','trips','activity_weighted_trips','council_weighted_trips','act_cncl_weighted_trips']],
#         on=['origin','destination'],
#         how='left',
#         suffixes=('_2019','_2020')
#         ).merge(
#         od_df2021[['origin','destination','trips','activity_weighted_trips','council_weighted_trips','act_cncl_weighted_trips']],
#         on=['origin','destination'],
#         how='left',
#         suffixes=('','_2021')
#         ).merge(
#         od_df2022[['origin','destination','trips','activity_weighted_trips','council_weighted_trips','act_cncl_weighted_trips']],
#         on=['origin','destination'],
#         how='left',
#         suffixes=('_2021','_2022')
#         )
#     )

#     avg_od['trips']=avg_od[['trips_2019','trips_2020','trips_2021','trips_2022']].mean(axis=1)
#     avg_od['activity_weighted_trips']=avg_od[['activity_weighted_trips_2019','activity_weighted_trips_2020','activity_weighted_trips_2021','activity_weighted_trips_2022']].mean(axis=1)
#     avg_od['council_weighted_trips']=avg_od[['council_weighted_trips_2019','council_weighted_trips_2020','council_weighted_trips_2021','council_weighted_trips_2022']].mean(axis=1)
#     avg_od['act_cncl_weighted_trips']=avg_od[['act_cncl_weighted_trips_2019','act_cncl_weighted_trips_2020','act_cncl_weighted_trips_2021','act_cncl_weighted_trips_2022']].mean(axis=1)

#     avg_od['perc_trips']=(avg_od['trips']/avg_od['trips'].sum())*100
#     avg_od['perc_activity_weighted_trips']=(avg_od['activity_weighted_trips']/avg_od['activity_weighted_trips'].sum())*100
#     avg_od['perc_council_weighted_trips']=(avg_od['council_weighted_trips']/avg_od['council_weighted_trips'].sum())*100
#     avg_od['perc_act_cncl_weighted_trips']=(avg_od['act_cncl_weighted_trips']/avg_od['act_cncl_weighted_trips'].sum())*100

#     avg_od.drop(
#         ['trips_2019','trips_2020','trips_2021','trips_2022',
#         'activity_weighted_trips_2019','activity_weighted_trips_2020','activity_weighted_trips_2021','activity_weighted_trips_2022',
#         'council_weighted_trips_2019','council_weighted_trips_2020','council_weighted_trips_2021','council_weighted_trips_2022',
#         'act_cncl_weighted_trips_2019','act_cncl_weighted_trips_2020','act_cncl_weighted_trips_2021','act_cncl_weighted_trips_2022'],
#         axis=1,
#         inplace=True
#         )
#     avg_od['origin']=avg_od['origin'].map(iz_dict)
#     avg_od['destination']=avg_od['destination'].map(iz_dict)

#     return avg_od

def getAvergedOD(od_df2021,od_df2022,iz_dict):
    
    avg_od=(od_df2021.merge(
        od_df2022[['origin','destination','trips','activity_weighted_trips','council_weighted_trips','act_cncl_weighted_trips']],
        on=['origin','destination'],
        how='left',
        suffixes=('_2021','_2022')
        )
    )

    avg_od['trips']=avg_od[['trips_2021','trips_2022']].mean(axis=1)
    avg_od['activity_weighted_trips']=avg_od[['activity_weighted_trips_2021','activity_weighted_trips_2022']].mean(axis=1)
    avg_od['council_weighted_trips']=avg_od[['council_weighted_trips_2021','council_weighted_trips_2022']].mean(axis=1)
    avg_od['act_cncl_weighted_trips']=avg_od[['act_cncl_weighted_trips_2021','act_cncl_weighted_trips_2022']].mean(axis=1)

    avg_od['perc_trips']=(avg_od['trips']/avg_od['trips'].sum())*100
    avg_od['perc_activity_weighted_trips']=(avg_od['activity_weighted_trips']/avg_od['activity_weighted_trips'].sum())*100
    avg_od['perc_council_weighted_trips']=(avg_od['council_weighted_trips']/avg_od['council_weighted_trips'].sum())*100
    avg_od['perc_act_cncl_weighted_trips']=(avg_od['act_cncl_weighted_trips']/avg_od['act_cncl_weighted_trips'].sum())*100

    # avg_od.drop(
    #     ['trips_2021','trips_2022',
    #     'activity_weighted_trips_2021','activity_weighted_trips_2022',
    #     'council_weighted_trips_2021','council_weighted_trips_2022',
    #     'act_cncl_weighted_trips_2021','act_cncl_weighted_trips_2022'],
    #     axis=1,
    #     inplace=True
    #     )

    avg_od=avg_od[['origin','destination','perc_act_cncl_weighted_trips']]
    avg_od['origin']=avg_od['origin'].map(iz_dict)
    avg_od['destination']=avg_od['destination'].map(iz_dict)

    return avg_od
    

    
    
```
